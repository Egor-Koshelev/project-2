```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/content/PostgreSQL.csv')
```

```python
display(df)
```

```python
df.groupby('delivery_person_age')['target'].mean().plot(kind = 'line', figsize = (10, 5))
plt.xlabel('Возраст курьера')
plt.ylabel('Время доставки, мин')
plt.title('Зависимость времени доставки от возраста курьера')
plt.grid(True)
plt.show()
```

```python
plt.figure (figsize = (10, 5))
plt.scatter(df['delivery_person_age'], df['target'])
plt.xlabel('Возраст курьера')
plt.ylabel('Время доставки, мин')
plt.title('Scatter Plot : Время доставки/возраста курьера')
plt.grid(True)
plt.show()
```
